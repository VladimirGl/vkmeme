```python
from keras.models import Sequential
from keras.layers import GlobalAveragePooling2D
from keras.preprocessing import image
from keras.applications.resnet50 import ResNet50
from keras.applications.resnet50 import preprocess_input
from tqdm import tqdm

import numpy as np
import matplotlib.pyplot as plt

import sklearn
import cv2

from glob import glob
```

```python
ACTIVATIONS_PATH = '/Users/vladgl/Downloads/vkhack/pic_activations/'
activations_lst = sorted(glob(ACTIVATIONS_PATH + '*'))
```

```python
model = ResNet50(weights='imagenet', include_top=False)

avg_model = Sequential()
avg_model.add(model)
avg_model.add(GlobalAveragePooling2D())

def get_activations(pic_path):
    img = image.load_img(pic_path, target_size=(224, 224))
    img_data = image.img_to_array(img)
    img_data = np.expand_dims(img_data, axis=0)
    img_data = preprocess_input(img_data)
    
    return avg_model.predict(img_data)[:, 0]
```

```python
xs = []
ys = []

inverse_dict = dict()

for i, activation in enumerate(activations_lst):
    inverse_dict[i] = activation
    activations = np.load(activation)
    activations_label = [i] * len(activations)
    
    xs.append(activations)
    ys.append(activations_label)
```

```python
def to_image_string(image_filepath):
    return open(image_filepath, 'r').read().encode('base64')

def from_base64(base64_data):
    nparr = np.fromstring(base64_data.decode('base64'), np.uint8)
    return cv2.imdecode(nparr, cv2.IMREAD_ANYCOLOR)
```

```python
def decode_as_image(encoded_image_string):
    decoded_image = base64.b64decode(encoded_image_string)
    return cv2.imdecode(np.frombuffer(decoded_image, dtype=np.uint8), flags=cv2.IMREAD_UNCHANGED)

def encode_image(image):
    _, buff = cv2.imencode('.jpg', image)
    return base64.b64encode(buff).decode('utf-8')
```
